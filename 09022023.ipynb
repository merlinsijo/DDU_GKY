### Type cassting with pandas

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Car_sales.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.info()
```

### Checking the contents in latest launch column

```python
df.Latest_Launch.unique()        
```

### Change Latest Year type into datetime type

```python
df['Latest_Launch']=pd.to_datetime(df.Latest_Launch)
df
```

```python
df['Year']=pd.DatetimeIndex(df.Latest_Launch).year
df['Month']=pd.DatetimeIndex(df.Latest_Launch).month
df['Day']=pd.DatetimeIndex(df.Latest_Launch).day
df
```

```python
df.Month.replace({
    1:"Januvary",
    2:"Februvary",
    3:"March",
    4:"April",
    5:"May",
    6:"June",
    7:"July",
    8:"Auguest",
    9:"September",
    10:"October",
    11:"November",
    12:"December"
},inplace=True)
df.head()
```

```python
df.Fuel_efficiency=df.Fuel_efficiency.astype(int)
df.head(10)
```

### Importing label encoder from sklearn

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le=LabelEncoder()
```

```python
df["Le_Vehicle"] = Le.fit_transform(df.Vehicle_type)
df["Le_Manufacture"]=Le.fit_transform(df.Manufacturer)
df["Le_Model"]=Le.fit_transform(df.Model)
```
