### Handling null values with windspeed.csv

```python
df2=pd.read_csv('Windspeed.csv')
df2
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

### filling common values to filling missing data

```python
df2.fillna(0)
```

### adding missing data to indiviual column

```python
df2.fillna({
    "Temperature":0,
    "Windspeed":5,
    "Status":"sunny"
    
})
```

### forward fill 

```python
df2.fillna(method="ffill")
```

### Backword fill

```python
df2.fillna(method="bfill")
```

## limiting backword fill /forward fill

```python
df2.fillna(method="ffill",limit=1)
```

### Filling with pandas object

```python
df2.fillna(df2.mean())
```

### filling for specific range

```python
df2.fillna(df2.mean()['Temperature':'Windspeed'])
```

### interpolate means that the relation

```python
df2.interpolate()
```

```python
df2.interpolate(limit=1,limit_direction="backward")
```

```python
df2.interpolate(limit=1,limit_direction="both",limit_area='inside')
```

```python
df2.interpolate(limit=1,limit_direction="both",limit_area='outside')
```
